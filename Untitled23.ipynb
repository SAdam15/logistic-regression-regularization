```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split

# Funkcja do rysowania powierzchni decyzyjnej
def plot_classification_surface(model, X, y, ax=None, title=None):
    if ax is None:
        ax = plt.gca()
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300),
                         np.linspace(y_min, y_max, 300))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
    ax.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolor='k')
    if title:
        ax.set_title(title)

# Dane do klasyfikacji
X, y = make_classification(n_samples=500, n_features=2, n_redundant=0,
                           n_informative=2, n_clusters_per_class=1,
                           random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Lista modeli z różnymi regularyzacjami
reg_configs = [
    ('L2, C=1.0', 'l2', 1.0),
    ('L2, C=0.1', 'l2', 0.1),
    ('L2, C=10', 'l2', 10),
    ('L1, C=1.0', 'l1', 1.0),
    ('L1, C=0.1', 'l1', 0.1),
    ('L1, C=10', 'l1', 10),
    ('ElasticNet, C=1.0', 'elasticnet', 1.0),
]

fig, axes = plt.subplots(3, 3, figsize=(18, 12))
axes = axes.ravel()

for i, (desc, penalty, C) in enumerate(reg_configs):
    if penalty == 'elasticnet':
        model = LogisticRegression(penalty=penalty, C=C, solver='saga',
                                   l1_ratio=0.5, max_iter=5000, random_state=42)
    else:
        solver = 'liblinear' if penalty == 'l1' else 'lbfgs'
        model = LogisticRegression(penalty=penalty, C=C, solver=solver,
                                   max_iter=5000, random_state=42)
        
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    f1_train = f1_score(y_train, y_train_pred)
    f1_test = f1_score(y_test, y_test_pred)
    
    plot_classification_surface(
        model, X_train, y_train, ax=axes[i],
        title=f"{desc}\nF1 train: {f1_train:.2f}, test: {f1_test:.2f}"
    )

# Ukrycie pustych wykresów
for j in range(i+1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```
